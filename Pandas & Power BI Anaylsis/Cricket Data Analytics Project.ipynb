# T20 WORLD CUP Data Cleaning And Processing

Processing Match Results

```python
import pandas as pd
import json
```

```python
with open("C:/Users/Public/Kiran Personal/Domain/DATA Engineering/1 Power BI Projects/Codebasic-Cricket Data Analytics Project/Details & Resources/t20_wc_match_results.json") as f:
    data = json.load(f)
    
df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.head(10)
```

```python
df_match.shape
```

Rename scorecard to matchid

```python
df_match.rename({"scorecard":"matchid"}, axis = 1, inplace=True)
df_match.head(5)
```

```python
match_id_dict = {}

for i, row in df_match.iterrows():
    key1 = row["team1"] + " Vs " + row["team2"]
    key2 = row["team2"] + " Vs " + row["team1"]
    
    match_id_dict[key1] = row["matchid"]
    match_id_dict[key2] = row["matchid"]
    
```

```python
df_match.to_csv("C:/Users/Public/Kiran Personal/Domain/DATA Engineering/1 Power BI Projects/Codebasic-Cricket Data Analytics Project/dim_match_summary.csv", index=False)
```

Batting Summary

```python
with open("C:/Users/Public/Kiran Personal/Domain/DATA Engineering/1 Power BI Projects/Codebasic-Cricket Data Analytics Project/Details & Resources/t20_wc_batting_summary.json") as f:
    data = json.load(f)
    
all_record = [] 

for rec in data:
    all_record.extend(rec['battingSummary'])
    
df_batting = pd.DataFrame(all_record) 
df_batting.head(10)
```

```python
df_batting['out/not_out'] = df_batting.dismissal.apply(lambda x: "out" if len(x)>0 else 'not_out' )
df_batting.head(10)
```

```python
df_batting["match_id"] = df_batting["match"].map(match_id_dict)
df_batting
```

```python
df_batting.drop(columns = ["dismissal"], inplace = True)
df_batting
```

```python
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace("â€", ""))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace("\xa0", ""))
df_batting
```

```python
df_batting.to_csv("C:/Users/Public/Kiran Personal/Domain/DATA Engineering/1 Power BI Projects/Codebasic-Cricket Data Analytics Project/fact_batting_summary.csv", index = False)
```
